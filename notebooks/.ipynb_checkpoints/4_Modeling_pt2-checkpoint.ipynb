## KMeans Clustering

Let's try an unsupervised model that can help determine who many clusters exist among the 11 different users listening habits by genre.  My goal here is to see how much overlap there is in this portion of the data. 

```python
X = df
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

```python
X.head()
```

```python
km = KMeans(n_clusters=10, random_state=42)
km.fit(X_sc)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
km.predict(X_sc)
```

```python
df['cluster'] = km.labels_
df.head()
```

```python
df.groupby('cluster').mean()
```

```python
visualizer = KElbowVisualizer(km, k=(2,30), timings=True)
visualizer.fit(df)
visualizer.show();
```

According to the graph above, the ideal number of clusters for this dataset is 8, with a  distortion score of 219,461.914. 8 clusters is chosen because that's when the distortion scores beings to flatten off slightly.  From the 11 users' data, 8 genre clusters were created which shows that most genre preferences are unique.

```python
group = df['cluster']
```

```python
hyp.plot(df, '.', title='KMeans Clusters', ndims=2, n_clusters=10);
```

The plot above shows that most of the clusters are grouped on the margins of the data set with only a few small ones in between. I wanted to group the data according the cluster groups created in the cells above but I wasn't able to do so.

If I continue to work on this project, which I plan to, I'll take a closer look at which songs fell into which clusters and the similarities they share.

```python
silhouette_score(X_sc, km.labels_)
```

A negative silhouette score demonstrates there is both weak cohension of local points to their own clusters and seperation between clusters.

```python
scores = []
for k in range(2, 30):
    cl = KMeans(n_clusters=k)
    cl.fit(X_sc)
    inertia = cl.inertia_
    sil = silhouette_score(X_sc, cl.labels_)
    scores.append([k, inertia, sil])
    
score_df = pd.DataFrame(scores)
score_df.columns = ['k', 'inertia', 'silhouette']
```

```python
score_df.head()
```

```python
plt.rcdefaults()
fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].plot(score_df.k, score_df.inertia, color='#5299D3')
axes[0].set_title('Inertia over k')
axes[1].plot(score_df.k, score_df.silhouette, color='#5299D3');
axes[1].set_title('Silhouette Score over k');
fig.suptitle('Inertia and Silhouette Scores', fontsize=18);

ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

fig.tight_layout();
```

## K Nearest Neighbors

My goal is to determine how many classes are created when using KNN on the genre features in the dataset.

```python
df = pd.read_csv('../data_sl/model_df.csv')
```

```python
df.drop(columns=['artistName', 'trackName'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
X = df.drop(columns=['target'])
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ss = StandardScaler()
ss.fit(X_train)
```

Since KNN measures distance, the data needs to be scaled.

```python
X_train_sc = ss.transform(X_train)
```

```python
X_test_sc = ss.transform(X_test)
```

```python
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

The highest baseline score is about 25%, which makes sense because Will had the largest dataset among the group of users.

```python
knn = KNeighborsClassifier()
```

```python
cross_val_score(knn, X_train_sc, y_train).mean()
```

Running KNN already increases the score to 35%. Let's use a pipeline and gridsearch to see if that score can improve even more.

```python
pipe = Pipeline([
    ('ss', StandardScaler()),
    ('knn', KNeighborsClassifier())
])
```

```python
pipe_params = {'ss__with_mean': [True, False], 
             'ss__with_std': [True, False],
             'knn__p': [1, 2], 
             'knn__weights': ['uniform', 'distance'],
             'knn__metric': ['euclidean', 'manhattan'],
             'knn__n_neighbors': [3, 5, 7, 9, 11]}
```

```python
pipe_gridsearch = GridSearchCV(pipe,
                                pipe_params,
                                cv=3,
                                verbose=1)
```

```python
pipe_gridsearch.fit(X_train, y_train);
```

```python
print('Best score:', pipe_gridsearch.best_score_)
print('Best params:', pipe_gridsearch.best_params_)
print('Best estimator:', pipe_gridsearch.best_estimator_)
```

```python
pipe_gridsearch.score(X_test, y_test)
```

The score improves to 39% by running the data through the gridsearch. It's a solid improvement but shows that the model will guess incorrectly well over half the time. This suggests the taste in genre among the users is somewhat distinct but still has a lot of overlap.

```python
k_scores = []
k_range = range(1,40)

for k in k_range:
    model = KNeighborsClassifier(n_neighbors=k).fit(X, y)
    y_pred = model.predict(X_test_sc)
    k_scores.append(metrics.accuracy_score(y_test, y_pred))
```

```python
knn_df = pd.concat([pd.DataFrame(k_range, columns = ['k_value']), pd.DataFrame(k_scores, columns = ['score'])], axis = 1)
```

```python
knn_df.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data = knn_df, x='score', y = 'k_value', color='#5299D3')
plt.title('Accuracy vs K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.show()
```
