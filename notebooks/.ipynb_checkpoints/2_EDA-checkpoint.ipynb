## Feature Engineering 

Since this data was collected from Spotify's API, it's already very well structured and doesn't require too much cleaning. That being said, I'm sure there is some feature engineering that can be performed done. 

```python
df.head()
```

```python
df = pd.read_csv('../data/complete_df.csv')
```

```python
z = df['popularity'].quantile(0.8)
df['popularity'] = df['popularity'].apply(lambda x: 1 if x >= z else 0)
```

Popularity is ranked between 0 - 100. To make it easier for the models to determine if a song is popular, I'll change that score to a 0 or 1, making it a 1 if it's in the 80% percentile or higher.

```python
df['release_date'] = [x.split('-')[0] for x in df['release_date']]
```

Changing the release date to the year will make it cleaner for the model to read.  There is definitely more feature engineering to be done to show what users have been listening to lately, but I elected to not focus on that for this project.  Definitely something I'd consider doing in the future though.

```python
df['total_msPlayed'] = df.groupby(['artistName', 'trackName', 'target'])['msPlayed'].transform('sum')
```

In the dataframe's current state, each song that has been listened to has it's own row alongside the time it was listened to. This creates a lot of duplicates rows that I won't need. However, I don't want to delete the msPlayed from those duplicates, so I will add the msPlayed by song, by target together and then delete the copies.

```python
df.drop(columns=['endTime', 'msPlayed', 'track_id', 'artist_id', 'id'], axis=1, inplace=True)
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

After deleting the duplicates, the dataframe shrinks to ~28k rows.

```python
df.to_csv('../data_sl/model_df.csv', index=False)
```

## Histogram Comparison

```python
features = [i for i in df.columns][2:15]
features
```

```python
plt.rcdefaults()
fig, axs = plt.subplots(figsize=(12, 12), nrows=4, ncols=3)
fig.subplots_adjust(wspace=0.5, hspace=0.5, top=1, bottom=0.2)

me = df[df['target'] == 0]
sami = df[df['target'] == 1]
will = df[df['target'] == 2]


fig.suptitle('Audio Features Comparison by Density', fontsize=18)

for col, ax in zip(features, axs.flatten()):
    ax.hist(will[col], bins=30, alpha=0.8, label='Will', color='#3a86ff', stacked=True)
    ax.hist(sami[col], bins=30, alpha=0.8, label='Sami', color='#ff006e', stacked=True)
    ax.hist(me[col], bins=30, alpha=0.8, label='Me', color='#ffbe0b', stacked=True)
    
    ax.set_title(col)
    ax.legend()

fig.tight_layout();
plt.show()
```

My goal with this project is to determine how unique each listener's taste in music is. The plot above looks at three users (instead of all eleven to make it more readable) data and compares the density of the audio features using a histogram.  

Some immediate takeaways are the Will is the most active Spotify user and I am the least. Sami enjoys higher energy songs with low acousticness, which makes sense because he enjoys EDM/electronic music. We all prefer major over minor songs (0=minor key and 1=major key), songs with less liveness, and low instrumentalness.  Will also has an affinity for low valence (Spotify's score for musical positiveness) music.  I reached out to him when putting this together to make sure he was feeling OK.

## Spider Graphs

Since the correlation plots above were pretty useless. I wanted to use spider plots so I could overlay each users taste in music according to a few of the audio features.

```python
audio_features = df[["danceability", "energy", "key", "loudness", "mode", "speechiness", "acousticness", "instrumentalness", 
                     "liveness", "valence", "tempo", "duration_ms", "time_signature", "target"]]
```

```python
audio_features.head()
```

```python
grouped_aud_feats = audio_features.groupby('target').agg(['mean'])
```

```python
grouped_aud_feats = pd.DataFrame(grouped_aud_feats)
```

```python
grouped_aud_feats.head(11)
```

I created a new dataframe that took the mean of each users' audio features. 

```python
min_max_scaler = MinMaxScaler()
```

Since some of the audio features metrics are on different scales, tempo and duration_ms to be specific, I scaled the data so it could all be captured in the same plot.

```python
grouped_aud_feats.loc[:] = min_max_scaler.fit_transform(grouped_aud_feats.loc[:])
```

```python
grouped_aud_feats.head()
```

```python
grouped_aud_feats.drop(columns=['key', 'mode', 'duration_ms', 'speechiness', 'liveness', 
                                'time_signature'], axis=1, inplace=True)
```

Spider graphs start to lose their effectiveness when you add too many features, so I dropped a few to make the shapes look better.

```python
grouped_aud_feats
```

```python
col_names = ["danceability", "energy", "loudness", "acousticness", 
             "instrumentalness", "valence", "tempo"]
col_names = [*col_names, col_names[0]]

me = grouped_aud_feats.iloc[0]
me = [*me, me[0]]

sami = grouped_aud_feats.iloc[1]
sami = [*sami, sami[0]]

will = grouped_aud_feats.iloc[2]
will = [*will, will[0]]

ben = grouped_aud_feats.iloc[3]
ben = [*ben, ben[0]]

amanda = grouped_aud_feats.iloc[4]
amanda = [*amanda, amanda[0]]

matt = grouped_aud_feats.iloc[5]
matt = [*matt, matt[0]]

bruhl = grouped_aud_feats.iloc[6]
bruhl = [*bruhl, bruhl[0]]

jenna = grouped_aud_feats.iloc[7]
jenna = [*jenna, jenna[0]]

riley = grouped_aud_feats.iloc[8]
riley = [*riley, riley[0]]

dan = grouped_aud_feats.iloc[9]
dan = [*dan, dan[0]]

cam = grouped_aud_feats.iloc[10]
cam = [*cam, cam[0]]


fig = go.Figure(
    data=[
        go.Scatterpolar(r=me, theta=col_names, fill='toself', name='Me'),
        go.Scatterpolar(r=sami, theta=col_names, fill='toself', name='Sami'),
        go.Scatterpolar(r=will, theta=col_names, fill='toself', name='Will'),
        go.Scatterpolar(r=ben, theta=col_names, fill='toself', name='Ben'),
        go.Scatterpolar(r=amanda, theta=col_names, fill='toself', name='Amanda'),
        go.Scatterpolar(r=matt, theta=col_names, fill='toself', name='Matt'),
        go.Scatterpolar(r=bruhl, theta=col_names, fill='toself', name='Bruhl'),
        go.Scatterpolar(r=jenna, theta=col_names, fill='toself', name='Jenna'),
        go.Scatterpolar(r=riley, theta=col_names, fill='toself', name='Riley'),
        go.Scatterpolar(r=dan, theta=col_names, fill='toself', name='Dan'),
        go.Scatterpolar(r=cam, theta=col_names, fill='toself', name='Cam')
    ],
    layout=go.Layout(
        title=go.layout.Title(text='Audio Features Comparison'),
        polar={'radialaxis': {'visible': True}},
        showlegend=True
    )
)

pyo.plot(fig)
```

## Sankey Diagram

Instead of simply listing the top 20 genres, I was curious to see which users were listening to which genres. The section below is my not entirely successfull attempt at that. Sankey plots are also constructed in a way that I had to do a lot data manipulation in order to put the data into the correct format.

```python
genre_df = df.drop(columns=["endTime", "artist_id", "id", "track_id", "msPlayed", "artistName", "trackName", "release_date", "popularity", "danceability", "energy", "key",
                            "loudness","mode","speechiness","acousticness", "instrumentalness","liveness","valence","tempo",
                            "duration_ms","time_signature", "alternative dance", "art pop", "dance pop","edm", "indie pop", 
                            "indie soul", "indietronica", "new rave","permanent wave", "post-teen pop", "total_msPlayed"], axis= 1)
```

```python
genre_df = genre_df.groupby('target').agg(['sum'])
```

I slimmed down the genres dataframe even more and summed all the observations of each genre by user.

```python
source = range(0,5)
source = list(itertools.chain.from_iterable(itertools.repeat(x, 20) for x in source))
```

I had to make a sequence of 0-4 (representing the 5 users) and repeat them 20 times so they could link to the 20 features.

```python
values = []
for index, row in genre_df.iterrows():
    values.append(list(row[0:21]))
```

```python
flatten = lambda t: [item for values in t for item in values]
```

```python
value = flatten(values)
```

Then, I had to turn all the valueas in the dataframe into one long list.

```python
target = list(range(5,25)) * 5
```

Then I had to represent all the targets from 5-25 so that all the sources would connect to them. Was all this work worth it, spoiler alert, it was not, but I'm glad I learned how to make a sankey plot and have a better idea of when to use them going forward.

```python
# data
label = ["Me", "Sami", "Will", "Ben", "Amanda", 'alternative r&b','alternative rock','chillwave','classic rock',
         'electronica','electropop','escape room','folk rock','freak folk','funk','hip hop',
         'indie folk','indie rock','mellow gold','modern rock','pop','pop rock','rock',
         'soft rock','soul']
source = source
target = target
value = value
# data to dict, dict to sankey
link = dict(source = source, target = target, value = value)
node = dict(label = label, pad = 35, thickness = 10)
data = go.Sankey(link = link, node = node)
```
